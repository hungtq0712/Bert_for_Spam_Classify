```python
from ipywidgets import widgets
from IPython.display import display
import tensorflow as tf
import tensorflow_hub as hub

model_path = "/content/drive/MyDrive/NLP/bert"
loaded_model = tf.keras.models.load_model(model_path)

def check_spam(content):
    predictions = loaded_model.predict([content])
    if predictions[0] > 0.5:
        return "This is a spam email!"
    else:
        return "This is not a spam email."

# Tạo một text box để nhập nội dung email
text_box = widgets.Textarea(
    value='',
    placeholder='Type email content here...',
    description='Email content:',
    disabled=False
)

# Tạo một nút để thực hiện phân loại
button = widgets.Button(description="Classify")

# Hàm xử lý sự kiện khi nút được nhấn
def on_button_clicked(b):
    content = text_box.value
    result = check_spam(content)
    print(result)

button.on_click(on_button_clicked)

# Hiển thị các widget
display(text_box)
display(button)
```
